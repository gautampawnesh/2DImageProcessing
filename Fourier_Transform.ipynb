```python
import cv2
import numpy as np 

img = cv2.imread("images/horizontal_lines.jpg", cv2.IMREAD_GRAYSCALE)

np.fft.restore_all()

f = np.fft.fft2(img)

magnitude_spectrum = 20*np.log(np.abs(f))
print(magnitude_spectrum.shape)
magnitude_spectrum_ = np.asarray(magnitude_spectrum, dtype=np.uint8)

cv2.imshow("magnitude spectrum", magnitude_spectrum)

#cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images/horizontal_lines.jpg',0)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()

cv2.waitKey() 
cv2.destroyAllWindows()
```

### FFT - Shift Theorem

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# remove below line 
np.fft.restore_all()

img_1 = cv2.imread('images/road1.jpg',0)
img_2 = cv2.imread('images/road1.jpg',0)

f_1 = np.fft.fft2(img_1)
f_2 = np.fft.fft2(img_1)
#print()
f_2_conj = np.ma.conjugate(f_2)
# cross power spectrum coeff 
cross_power_coeff = (f_1*f_2_conj)/(np.abs(f_1*f_2_conj))
# inverse fft of cross_power coeff
r = np.fft.ifft2(cross_power_coeff).real

plt.subplot(121),plt.imshow(img_1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_2, cmap = 'gray')
plt.title('shifted ImAGE'), plt.xticks([]), plt.yticks([])
plt.show()
"""
find the place where r = 1 ;
"""
r = r.reshape(1, -1)
print(np.where(r == 1))
#cv2.imshow("phase correlation", r)
cv2.waitKey() 
cv2.destroyAllWindows()
```
