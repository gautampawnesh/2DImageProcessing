## Phase correlation: 

cross-power spectrum of image f1 and f2  and their Fourier transformation is denoted by F and F' :

$R\ =\ \frac{F\ \%*\%\ F^{'*}}{|FF^{'*} |}$

%*% --> Hadamard (entry-wise} product 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# remove below line 
np.fft.restore_all()

img_1 = cv2.imread('images/lamp.jpg',0)
img_2 = cv2.imread('images/shifted_lamp.jpg',0)

f_1 = np.fft.fft2(img_1)
f_2 = np.fft.fft2(img_2)
#print()
f_2_conj = np.ma.conjugate(f_2)
# cross power spectrum coeff 
cross_power_coeff = (f_1*f_2_conj)/(np.abs(f_1*f_2_conj))
# inverse fft of cross_power coeff
r = np.fft.ifft2(cross_power_coeff).real

plt.subplot(131),plt.imshow(img_1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_2, cmap = 'gray')
plt.title('shifted ImAGE'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(r, cmap = 'gray')
plt.title('phase correlation'), plt.xticks([]), plt.yticks([])
plt.show()
"""
find the place where r is maximum ;
"""
r = r.reshape(1, -1)
print("unit impulse point located at :")
print(np.where(r == np.amax(r) ))
cv2.waitKey() 
cv2.destroyAllWindows()
```

\textbf{Algorithm for Phase correlation:}



1. Given two images $\displaystyle f_{1}$ and $\displaystyle f_{2}$ . 

2. Calculate discrete fourier transformation of both $\displaystyle f_{1}$ and $\displaystyle f_{2}$

				$\displaystyle  \begin{array}{{>{\displaystyle}l}}
F\ =\ F( f_{1}) \ \\
F^{'} \ =\ F( f_{2})
\end{array}$

3. Calculate the cross power spectrum by taking the complex conjugate of $\displaystyle F^{'}$, multiplying the fourier transformations together elementwise and normalising the product elementwise.



					$\displaystyle R\ =\ \frac{F\ \%*\%\ F^{'*}}{|FF^{'*} |}$

4. Obtain normalised cross-correlation by applying the inverse fourier transformation.



						$\displaystyle r\ =\ F^{-1}\{R\}$	

5. Determine the location, where r is maximum:

				 $\displaystyle ( \Delta x,\ \Delta y) \ =\ argmax_{x,y} \ \{r\}$
